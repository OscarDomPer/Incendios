```python
# Inicializar el geolocalizador
geolocator = Nominatim(user_agent="prediccion_incendios")

# Definir la función para obtener la ubicación
def get_location(row):
    location = geolocator.reverse(f"{row['Latitud']}, {row['Longitud']}")
    return location

# Aplicar la función para obtener la información de ubicación
tqdm.pandas()
df['location_info'] = df.progress_apply(get_location, axis=1)

# Definir funciones para obtener provincia y municipio
def obtener_provincia(loc):
    if loc is not None and loc.raw.get('address'):
        return loc.raw.get('address', {}).get('state', '')
    else:
        return ''

def obtener_municipio(loc):
    if loc is not None and loc.raw.get('address'):
        return loc.raw.get('address', {}).get('city', '')
    else:
        return ''

# Aplicar las funciones para obtener provincia y municipio
df['provincia'] = df['location_info'].apply(obtener_provincia)
df['municipio'] = df['location_info'].apply(obtener_municipio)

# Mostrar el DataFrame resultante
print(df)

df.to_csv('Data/data_Galicia/intermedios/neg_gal.csv', index=False)
```

Añadir la estación metereológica más cercana a cada punto

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="prediccion_incendios")

lugares = ["Val do Dubra, A Coruña, Spain", "Boiro, A Coruña, Spain","Lousame, A Coruña, Spain","Narón, A Coruña, Spain", 
           "Castrelo de Miño, Ourense, Spain", "A Veiga, Ourense, Spain", "Calvos de Randin, Ourense, Spain", "Ríos, Ourense, Spain",
           "Baleira, Lugo, Spain", "Mondoñedo, Lugo, Spain", "Antas de Ulla, Lugo, Spain", "Chantada, Lugo, Spain", 
            "Ponteareas, Pontevedra, Spain", "Meis, Pontevedra, Spain",  "O Rosal, Pontevedra, Spain", "Lalín, Pontevedra, Spain"]

coordenadas = {}

for lugar in lugares:
    # Dividir la cadena en partes separadas por comas y obtener el primer elemento (el nombre del pueblo)
    nombre_pueblo = lugar.split(',')[0]
    
    location = geolocator.geocode(lugar)
    if location:
        # Usar solo el nombre del pueblo como clave en el diccionario
        coordenadas[nombre_pueblo] = (location.latitude, location.longitude)

# Mostrar el diccionario con los nombres de los pueblos y sus coordenadas
print(coordenadas)
```

```python
from geopy.distance import geodesic

# Crear una función para encontrar la estación más cercana
def estacion_mas_cercana(row):
    distancias = {}
    for lugar, coordenada in coordenadas.items():
        distancias[lugar] = geodesic((row['latitude'], row['longitude']), coordenada).kilometers
    estacion_cercana = min(distancias, key=distancias.get)
    return estacion_cercana

# Aplicar la función a cada fila del DataFrame para encontrar la estación más cercana
df['Estacion'] = ''

for idx, row in tqdm(df.iterrows(), total=len(df)):
    df.at[idx, 'Estacion'] = estacion_mas_cercana(row)

# Guardar el DataFrame en un archivo sobrescribiendo si ya existe
archivo_salida = 'Data/data_Galicia/intermedios/neg_gal_prov.csv'  # Reemplaza 'nombre_del_archivo.csv' con el nombre de tu archivo de salida
df.to_csv(archivo_salida, index=False, mode='w')

print(f"Archivo guardado en: {archivo_salida}")
```

```python
# Definir la función para generar fechas aleatorias
def generar_fechas_aleatorias(inicio, fin, cantidad):
    inicio = datetime.strptime(inicio, '%Y-%m-%d')
    fin = datetime.strptime(fin, '%Y-%m-%d')
    delta = fin - inicio
    fechas_aleatorias = [inicio + timedelta(days=random.randint(0, delta.days)) for _ in range(cantidad)]
    return fechas_aleatorias

# Generar fechas aleatorias
fechas_aleatorias = generar_fechas_aleatorias('2008-01-01', '2023-12-31', len(df))

# Asignar las fechas aleatorias al DataFrame
df['acq_date'] = fechas_aleatorias

# Verificar el resultado

df.head()
```

```python

# Función para generar horas y minutos aleatorios
def generar_horas_aleatorias(cantidad):
    horas_aleatorias = [f"{random.randint(0, 23):02d}{random.randint(0, 59):02d}" for _ in range(cantidad)]
    return horas_aleatorias

# Generar horas y minutos aleatorios
horas_aleatorias = generar_horas_aleatorias(len(df))

# Asignar las horas y minutos aleatorios al DataFrame
df['acq_time'] = horas_aleatorias

# Verificar el resultado
df.head()
```

```python
import pandas as pd
import unicodedata
from tqdm.auto import tqdm

def normalizar_texto(texto):
    """Normaliza cadenas de texto eliminando tildes y caracteres especiales."""
    return unicodedata.normalize('NFKD', texto).encode('ascii', 'ignore').decode('ascii').lower()

def procesar_archivos(archivo_incendios, archivo_clima, archivo_salida):
    # Cargar los datos
    df_incendios = pd.read_csv(archivo_incendios)
    df_clima = pd.read_csv(archivo_clima)

    # Convertir las fechas a formato de fecha
    df_incendios['acq_date'] = pd.to_datetime(df_incendios['acq_date'])
    df_clima['Fecha'] = pd.to_datetime(df_clima['Fecha'])

    # Normalizar nombres de estaciones
    df_incendios['Estacion'] = df_incendios['Estacion'].apply(normalizar_texto)
    df_clima['Estacion'] = df_clima['Estacion'].apply(normalizar_texto)

    # Fusionar los datasets
    print("Combinando datasets...")
    df_combinado = pd.merge(df_incendios, df_clima, left_on=['acq_date', 'Estacion'], right_on=['Fecha', 'Estacion'], how='left')

    # Eliminar columnas redundantes y renombrar según sea necesario
    df_combinado.drop(columns=['Fecha_y'], inplace=True, errors='ignore')
    df_combinado.rename(columns={'Fecha_x': 'Fecha'}, inplace=True)

    # Guardar el archivo combinado
    df_combinado.to_csv(archivo_salida, index=False)
    print(f"Archivo combinado guardado en: {archivo_salida}")
    
    return df_combinado  # Retornar el DataFrame combinado

# Rutas de los archivos
archivo_incendios = 'Data/data_Galicia/intermedios/gal_neg_est.csv'
archivo_clima = 'Data/data_Galicia/clima_estaciones.csv'
archivo_salida = 'Data/data_Galicia/incen_gal_neg_est.csv'

# Ejecutar la función y capturar el DataFrame resultante
df_combinado = procesar_archivos(archivo_incendios, archivo_clima, archivo_salida)

# Agregar un print adicional
print("Proceso completado.")

# Imprimir el dataset resultante
print("Dataset resultante:")
df_combinado
```
